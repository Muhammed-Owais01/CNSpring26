```python
import socket
import threading
```

```python
s = socket.socket()
s.connect(('localhost', 9999))
```

```python
def receive():
    while True:
        try: print(s.recv(1024).decode())
        except: break

threading.Thread(target=receive).start()
```

```python
while True:
    msg = input("Enter a message: ")
    s.send(msg.encode())
```

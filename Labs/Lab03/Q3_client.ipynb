```python
import socket
import os
import threading
```

```python
s = socket.socket()
s.connect(('localhost', 9999))
```

```python
def receive():
    while True:
        try: print(s.recv(1024).decode())
        except: break

threading.Thread(target=receive).start()
```

```python
while True:
    msg = input("Enter message or /file filename: ")

    if msg.startswith("/file"):
        filename = msg.split()[1]

        s.send(f"file:{filename}".encode())

        reply = s.recv(1024).decode()
        print(reply)

        if "accepted" in reply:
            with open(filename, "rb") as f:
                s.send(f.read())
    else:
        s.send(msg.encode())
```
